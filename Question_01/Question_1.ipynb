```python
data = pd.read_csv("C:\\Users\\stell\\Documents\\GitHub\\Python\\Assessment_Python_DVDRental\\Query_01\\Query_1_SQL.csv")
```

```python
data['payment_date'] = pd.to_datetime(data['payment_date'])
```

```python
data['payment_date'] = data['payment_date'].dt.date
```

```python
grouped_data = data.groupby(['country', 'payment_date']).sum().reset_index()
```

```python
countries = ['China', 'Germany', 'India']
filtered_data = grouped_data[grouped_data['country'].isin(countries)]
```

```python
pivoted_data = filtered_data.pivot(index='payment_date', columns='country', values='total_amount')
```

```python
plt.figure(figsize=(10, 6))
for country in countries:
    plt.plot(pivoted_data.index, pivoted_data[country], label=country)
plt.xlabel('Payment Date')
plt.ylabel('Total Revenue')
plt.title('Daily Total Revenue for Three Countries')
plt.legend()
plt.show()
```
