```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Query_03_df = pd.read_csv("C:\\Users\\stell\\Documents\\GitHub\\Python\\Assessment_Python_DVDRental\\Query_03\\Query_3_SQL.csv")
Query_03_df.head()
```

```python
rev = pd.DataFrame(Query_03_df, columns=['country', 'total_amount'])
print(rev)
```

```python
countries = rev['country']
revenue = rev['total_amount']
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(countries, revenue)
ax.set_xlabel('Countries')
ax.set_ylabel('Total Revenue')
ax.set_title('Total Amount per Country')
plt.xticks(rotation=90)
plt.show()
```
